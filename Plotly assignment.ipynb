```python
import plotly.express as px
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'titanic.csv'
data_titanic=pd.read_csv(file_path)
data_titanic
```

```python
data_titanic.dropna(subset=["Cabin"],inplace =True)
data_titanic
```

```python
data_titanic.dropna(subset=["Embarked"],inplace =True)
data_titanic
```

```python
embarked_count = data_titanic.groupby('Embarked')['PassengerId'].count()
embarked_count
fig=px.bar( data_titanic, x= 'Embarked',y= Passenger count'
```

```python
mendata = data_titanic[data_titanic.Sex == 'male']
womendata = data_titanic[data_titanic.Sex == 'female']
print("No of males: " + str(mendata.count()['Sex']))
print("No of females: " + str(womendata.count()['Sex']))

```

```python
fig = px.bar(data_titanic, x='Sex', y='Survived', color='Survived', text='Survived', barmode='group', 
             labels={'Survived': 'Number of Passengers'},
             title='Survival Rate by Gender')
fig.show()
```

```python
fig = px.histogram (data_titanic, x= 'Age', color = 'Survived', nbins=30,
                    labels={'Age': 'Age of Passengers', 'count': 'Number of passengers'},
                    title ='Age distribution of Passengers by survival')
fig.show()
                    
```

```python
fig = px.violin(data_titanic, x='Pclass', y='Fare',
                labels ={'Pclass': 'Passenger class', 'Fare':'Fare price'},
                title='Fare prices')
fig.show()
```

```python
fig=px.scatter(data_titanic, x= 'Age',y='Fare',color='Pclass',
               labels={'Age': 'Age of passengers', 'Fare': 'Fare price'},
               title='Relationship between Passengers Age vs Fare Price')
fig.show()
```

```python
pd.DataFrame.iteritems = pd.DataFrame.items
fig=px.scatter_matrix (data_titanic,dimensions=['Age','Pclass'], color='Survived',
                      labels={'Age': 'Age of passegers', 'Fare':'Fare price', 'Pclass': 'Passenger class','Survived': 'Survival Status'},
                      title='Scatter matrix for Age,Fare and Pclass')
fig.show()
```
